```python
import random
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs, make_moons
from sklearn.cluster import KMeans, DBSCAN
```

```python
x, y = make_blobs(n_samples=100, centers=3, n_features=2,random_state=170)
```

```python
y_pred = KMeans(n_clusters=2, random_state=170).fit_predict(x)

plt.subplot()
plt.scatter(x[:, 0], x[:, 1], c=y_pred)
plt.title("Incorrect Number of Blobs")
```

```python
y_pred = KMeans(n_clusters=3, random_state=170).fit_predict(x)

plt.subplot()
plt.scatter(x[:, 0], x[:, 1], c=y_pred)
plt.title("Correct Number of Blobs")
```

```python
y_pred = DBSCAN(eps=3, min_samples=2).fit_predict(x)
plt.subplot()
plt.scatter(x[:, 0], x[:, 1], c=y_pred)
plt.title("Correct Clustering")
```

```python
t = np.linspace(0, 4*np.pi, 500)
f = 1.15
data = 3.0*np.sin(f*t + 0.001) + 0.5 + np.random.randn(500)

plt.scatter(t, data)
plt.title("Sin Points")
```

```python
data=data.reshape(-1,1)
```

```python
y_pred = DBSCAN(eps=2, min_samples=2).fit_predict(data)
plt.subplot()
plt.scatter(t, data[:, 0], c=y_pred)
plt.title("Incorrect hyperparameters")
```

```python
y_pred = KMeans(n_clusters=3, random_state=170).fit_predict(data)

plt.subplot()
plt.scatter(t, data[:, 0], c=y_pred)
plt.title("Kmeans-Incorrect Number of clusters")
```

```python
y_pred = KMeans(n_clusters=5, random_state=170).fit_predict(data)

plt.subplot()
plt.scatter(t, data[:, 0], c=y_pred)
plt.title("Kmeans-Incorrect Number of clusters") 
```

```python
x,y = make_moons(n_samples=1000, noise=0.05)
```

```python
y_pred = KMeans(n_clusters=2, random_state=170).fit_predict(x)

plt.subplot()
plt.scatter(x[:,0], x[:, 1], c=y_pred)
plt.title("Kmeans-Incorrect clustering") 
```

```python
y_pred = DBSCAN(eps=0.1, min_samples=10).fit_predict(x)

plt.subplot()
plt.scatter(x[:,0], x[:, 1], c=y_pred)
plt.title("DBSCAN-Moons")
```
